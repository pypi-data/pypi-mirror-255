```python
import numpy as np
import matplotlib.pyplot as plt
import spiakid_simulation.PhotonSimulator as PS
```

```python
test = PS.Photon_simulator('Template_Gaussian.yaml')
```

```python
x = []
y = []
lbd = []
for i in range(len(test.photon_dict)):
    x.append(test.photon_dict[i][0])
    y.append(test.photon_dict[i][1])
    lbd.append(test.photon_dict[i][2])
```

```python
plt.scatter(x,y,c=lbd)
plt.colorbar()
pos = [test.star_pos[0][1],test.star_pos[0][2]]
plt.title('Photon of a star at '+str(pos))
```
